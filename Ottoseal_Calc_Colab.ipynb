```python
# Импорт необходимых библиотек
import ipywidgets as widgets
from IPython.display import display, clear_output
from math import ceil

# Создание виджетов для ввода параметров
length_input = widgets.IntText(description='Длина (мм):', value=5000)
width_input = widgets.IntText(description='Ширина (мм):', value=10)
depth_input = widgets.IntText(description='Глубина (мм):', value=5)

package_type = widgets.Dropdown(
    options=['Без упаковки', 'Картридж 310 мл', 'Фольгированная упаковка 400 мл', 'Фольгированная упаковка 580 мл'],
    description='Упаковка:'
)

calc_button = widgets.Button(description='Рассчитать', button_style='success')

output = widgets.Output()

# Функция расчёта герметика
def calculate_sealant(b):
    with output:
        clear_output()

        # Получение данных из полей ввода
        length_m = length_input.value / 1000  # перевод длины в метры
        width_mm = width_input.value
        depth_mm = depth_input.value

        # Расчёт объёма герметика
        volume_ml = length_m * width_mm * depth_mm

        # Учёт коэффициента потерь 15%
        volume_with_loss_ml = volume_ml * 1.15

        # Вывод результатов
        print(f'Расчётный объём без потерь: {volume_ml:.2f} мл')
        print(f'Расчётный объём с учётом потерь (15%): {volume_with_loss_ml:.2f} мл')

        # Расчёт количества упаковок (если выбрана упаковка)
        if package_type.value != 'Без упаковки':
            package_volumes = {
                'Картридж 310 мл': 310,
                'Фольгированная упаковка 400 мл': 400,
                'Фольгированная упаковка 580 мл': 580
            }

            package_volume = package_volumes[package_type.value]
            packages_needed = ceil(volume_with_loss_ml / package_volume)

            print(f'Необходимое количество упаковок ({package_type.value}): {packages_needed} шт.')

# Привязка функции к кнопке
calc_button.on_click(calculate_sealant)

# Отображение виджетов интерфейса
widgets.VBox([
    length_input,
    width_input,
    depth_input,
    package_type,
    calc_button,
    output
])

# Варианты встраивания калькулятора на сайт:
# 1. Опубликуйте данный Colab-ноутбук (меню «Файл» → «Опубликовать в интернете») и затем встроите его на сайт с помощью HTML-тега <iframe>.
# 2. Используйте сервисы Binder или Voila для создания отдельного веб-приложения, встраиваемого через iframe.
```
